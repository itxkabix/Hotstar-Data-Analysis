```python
import pandas as pd #Data Manipulation
import matplotlib.pyplot as plt #Data Visualization
import seaborn as sns #Data Visualization

#as is used to define alias
#run shortcut - ctrl+enter
```

```python
df = pd.read_csv("hotstar.csv") #load your data in variable df
df
```

```python
df.head() #by default display first 5 rows
```

```python
df.tail() #by default display last 5 rows
```

```python
df.info() #gives basic info about the dataset
```

```python
df.describe()
```

```python
#to find total null values in our dataset
df.isnull().sum()
```

```python
#mean, median - replace numberic values (prefer median if data has outliers)
#mode, ffill, bfill - replace categorical data

#fillna() - is used to fill null values

df['running_time'] = df["running_time"].fillna(df['running_time'].median())
```

```python
#check for duplicates

df.duplicated().sum()

#if duplicates are there then use
df = df.drop_duplicates()
```

```python
#to visualize count of tv shows and movies

sns.countplot(x = df["type"], data = df, color = "coral") #to count the number of times each value in type column is appearing
plt.title("Count of Movies and TV Shows") #this gives title to our chart
plt.show() #to display the chart
```

```python
gb = df.groupby("type").agg({"hotstar_id":"count"}) #similar to pivots in excel
gb
```

```python
plt.figure(figsize = (10,4)) #to incread the width and height of chart
sns.countplot(x = df["genre"], data = df, color = "coral")
plt.title("Count of Genres")
plt.xticks(rotation = 90) #to rotate the tick labels on x-axis
plt.show()
```

```python
gb = df.groupby("age_rating").agg({"hotstar_id":"count"})
gb
```

```python
plt.pie(gb["hotstar_id"], labels = gb.index, autopct = "%.2f%%")
plt.show()
```

```python
gb = df.groupby("year").agg({"hotstar_id":"count"})
gb
```

```python
plt.figure(figsize = (10,4))
sns.lineplot(x = gb.index, y = gb["hotstar_id"])
plt.show()
```
